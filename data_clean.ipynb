```python
import pandas as pd 
import numpy as np
```

```python
# read the data 
data = pd.read_csv("SummerStudentAdmissions2.csv")
data.head()
```

## clean the data 

* Decision: only remains "Admit", "Waitlist", "Decline"

* State: ... 

* GPA: default to 4.0 scale 

* WorkExp: remove anomalies 

* TestScore: ...

* WritingScore: ...

* Gender: remove anomalies 

* VolunteerLevel: ...


```python
data = data.loc[(data['Decision'] == 'Admit') | (data['Decision'] == 'Waitlist') | (data['Decision'] == 'Decline')]
```

```python
data["GPA"] = np.where(data["GPA"]>4.0, np.nan, data["GPA"])
```

```python
data["WorkExp"] = np.where(data["WorkExp"]>10, np.nan, data["WorkExp"])
```

```python
data = data.loc[(data['Gender'] == 1.0) | (data['Gender'] == 0.0)]
```

```python
data = data.dropna()
```
